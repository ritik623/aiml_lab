```python
import pandas as pd
```

```python
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
import io
```

```python
 df = pd.read_csv(io.BytesIO(uploaded['attendance.csv'])) #reading a csv file
```

```python
df.head(7) #to read first 7 elments
```

```python
df.isnull() # to check wheter there is a null value or not represented by True or False
```

```python
df.isnull().sum() # to show the sum of the null value
```

```python
df.describe() # to count the mean, max and min value of the given csv file
```

```python
df.tail(7) # to print last 7 elemnts of the csv file
```

```python
months = ['Jan', 'Feb', 'March', 'Apr', 'May']# Convert percentage strings to float values
for month in months:
    df[month] = df[month].astype(str).str.rstrip('%').astype('float') / 100.0
df['Avg'] = (df[months].mean(axis=1))*100 # To find the avg value entered in the present csv file
```

```python
# Print monthly average percentages
print("Monthly Average Percentages:")
print(df[['SAP ID','ROLL ID', 'NAME', 'Avg']])
```

```python
df.min()# To find the minimum value present in the csv file
```

```python
df.max()# To find the maximum value present in the csv file
```
